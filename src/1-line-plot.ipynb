### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y = x**2
plt.plot(x, y, label="y = x^2", color='b', linestyle='-', linewidth=2)
plt.title("Line Plot of y = x^2")  
plt.xlabel("X Values")  
plt.ylabel("Y Values (x^2)")
plt.grid(True)
plt.legend() 
```
