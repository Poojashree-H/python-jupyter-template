### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Data for the plots
x = np.linspace(0, 10, 100)
y = np.sin(x)

bar_labels = ['A', 'B', 'C', 'D']
bar_values = [10, 20, 15, 30]

scatter_x = np.random.rand(50)
scatter_y = np.random.rand(50)

pie_labels = ['Apple', 'Banana', 'Cherry', 'Date']
pie_sizes = [25, 35, 20, 20]

# Step 2: Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# First subplot: Line chart
axs[0, 0].plot(x, y, label='y = sin(x)', color='b')
axs[0, 0].set_title("Line Chart")
axs[0, 0].set_xlabel("X Values")
axs[0, 0].set_ylabel("Y Values")
axs[0, 0].grid(True)
axs[0, 0].legend()

# Second subplot: Bar chart
axs[0, 1].bar(bar_labels, bar_values, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")

# Third subplot: Scatter plot
axs[1, 0].scatter(scatter_x, scatter_y, color='purple')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("X Values")
axs[1, 0].set_ylabel("Y Values")

# Fourth subplot: Pie chart
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
axs[1, 1].set_title("Pie Chart")

# Step 3: Adjust layout for better spacing
plt.tight_layout()

# Step 4: Show the plot
plt.show()
```
