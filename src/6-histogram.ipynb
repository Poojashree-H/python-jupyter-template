### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Generate random data for the histogram
data = np.random.randn(1000)  # 1000 random values from a normal distribution

# Step 2: Create the histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)

# Step 3: Customize the plot
plt.title("Customized Histogram")  # Add a title
plt.xlabel("Data Values")  # Label for the x-axis
plt.ylabel("Frequency")  # Label for the y-axis
plt.grid(True)  # Add grid
plt.xticks(np.arange(-4, 5, 1))  # Custom x-axis ticks
plt.yticks(np.arange(0, 200, 20))  # Custom y-axis ticks

# Step 4: Display the plot
plt.show()
```
